```python
import tensorflow as tf

dataset = tf.keras.utils.image_dataset_from_directory(
    '/content/drive/MyDrive/Colab Notebooks/Fish Data',
    image_size=(224, 224),
    batch_size=32,
    shuffle=True
)

# Accessing class names
print(dataset.class_names)
```

EDA

```python
import matplotlib.pyplot as plt
import numpy as np

# Preview one batch of images
class_names = dataset.class_names

plt.figure(figsize=(10, 10))
for images, labels in dataset.take(1):
    for i in range(9):
        ax = plt.subplot(3, 3, i + 1)
        plt.imshow(images[i].numpy().astype("uint8"))
        plt.title(class_names[labels[i]])
        plt.axis("off")
```

```python
for images, _ in dataset.take(1):
    print("Image shape:", images[0].shape)
    print("Image dtype:", images[0].dtype)
```

```python
image_min = tf.reduce_min(images).numpy()
image_max = tf.reduce_max(images).numpy()
print(f"Pixel values range from {image_min} to {image_max}")
```
